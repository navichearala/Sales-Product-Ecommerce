```python
import pandas as pd

data = pd.read_csv("Sales Product Ecommerce.csv")
print(data.head()) 
```

```python
print(data.isnull().sum())
#checking missing values
```

```python
data['Price Each'] = pd.to_numeric(data['Price Each'], errors='coerce')
data['Discount'] = data['Discount'].str.replace(',', '.').astype(float)
# Convert 'Price Each' and 'Discount' to numeric (replace commas with dots in 'Discount')
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')
# Convert 'Order Date' to datetime
```

```python
data.drop_duplicates(inplace=True)
# Remove duplicates
```

```python
print(data.isnull().sum())
# Check for null values after conversions
```

```python
null_rows = data[data['Price Each'].isnull()]
print(null_rows.head())
```

```python
# Fill missing 'Price Each' values with product-wise mean
data['Price Each'] = data.groupby('Product')['Price Each'].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
print(data.isnull().sum())
#dataset looks clean now, as there are no null values in any column.
```

```python
null_rows = data[data['Price Each'].isnull()]
print(null_rows.head())
```

```python
#Analysis
```

```python
data['Total Sales'] = data['Price Each'] * data['Quantity Ordered']
```

```python
# EDA by top salse 
top_products = data.groupby('Product')['Total Sales'].sum().sort_values(ascending=False)
print(top_products.head())
```

```python
#Sales by Category Analyze sales distribution by product category:
category_sales = data.groupby('Category')['Total Sales'].sum().sort_values(ascending=False)
print(category_sales)
```

```python
#City-Wise Performance Check which cities had the highest sales:
city_sales = data.groupby('City Store')['Total Sales'].sum().sort_values(ascending=False)
print(city_sales)
```

```python
#Monthly Sales Trends Extract the month from the Order Date column and analyze monthly trends:
data['Order Month'] = data['Order Date'].dt.month
monthly_sales = data.groupby('Order Month')['Total Sales'].sum()
print(monthly_sales)
```

```python
#KPI Analysis
```

```python
total_revenue = data['Total Sales'].sum()
print("Total Revenue:", total_revenue)
```

```python
#average order value
total_orders = data['Order ID'].nunique()
average_order_value = total_revenue / total_orders
print("Average Order Value (AOV):", average_order_value)
```

```python
#revenu by product
revenue_by_product = data.groupby('Product')['Total Sales'].sum()
print(revenue_by_product)
```

```python
#Top 10% Products by Revenue

cumulative_revenue = revenue_by_product.cumsum()
total_revenue = revenue_by_product.sum()
top_10_percent_products = cumulative_revenue[cumulative_revenue <= total_revenue * 0.10]
print("Top 10% Products by Revenue:")
print(top_10_percent_products)
```

```python
#Customer Lifetime Value (CLV)

purchase_frequency = total_orders / data['Customer Shipping Address'].nunique()
average_customer_lifespan = 3  # Example value, replace with actual
clv = average_order_value * purchase_frequency * average_customer_lifespan
print("Customer Lifetime Value (CLV):", clv)
```

```python
#Monthly Growth Rate

monthly_growth = monthly_sales.pct_change() * 100
print("Monthly Growth Rate (%):")
print(monthly_growth)
```

```python
#Profit Margin

total_costs = data['Price Each'].sum()  # Replace 'Cost' with actual column name if available
profit_margin = ((total_revenue - total_costs) / total_revenue) * 100
print("Profit Margin (%):", profit_margin)
```

```python
from fpdf import FPDF
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Load your cleaned dataset
data = pd.read_csv("Sales Product Ecommerce.csv")

# Generate some key metrics
total_revenue = data['Revenue'].sum()
aov = data['Revenue'].mean()
monthly_revenue = data.groupby(data['Order Date'].dt.to_period('M'))['Revenue'].sum()

# Generate visualizations dynamically
plt.figure(figsize=(10, 6))
monthly_revenue.plot(marker='o')
plt.title("Monthly Revenue Trends")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid()
plt.savefig("monthly_revenue_trends.png")  # Save the plot as an image
plt.close()

# Step 2: Create a PDF report using FPDF
class PDF(FPDF):
    def header(self):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, "Sales Analysis Report", align='C', ln=True)
        self.ln(10)

    def footer(self):
        self.set_y(-15)
        self.set_font('Arial', 'I', 8)
        self.cell(0, 10, f'Page {self.page_no()}', align='C')

# Initialize the PDF
pdf = PDF()
pdf.add_page()

# Add content
pdf.set_font('Arial', '', 12)
pdf.cell(0, 10, f"Total Revenue: ${total_revenue:,.2f}", ln=True)
pdf.cell(0, 10, f"Average Order Value (AOV): ${aov:,.2f}", ln=True)

# Add visualizations
pdf.ln(10)
pdf.cell(0, 10, "Monthly Revenue Trends:", ln=True)
pdf.image("monthly_revenue_trends.png", x=10, y=None, w=190)

# Save the PDF
pdf.output("Sales_Report.pdf")
print("PDF Report generated successfully!")
```
